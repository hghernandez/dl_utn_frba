```python
import tensorflow as tf
from tensorflow import keras


new_model = keras.models.load_model('ModelProd.PrimerNN')
```

```python
#Cargamos el nÃºmero 0_0

import cv2

imagen = cv2.imread('Numeros/0_0.png', 0)    # El 0 indica que se lee la imagen>
imagen = imagen.flatten().astype("float32") / 255.0
print(imagen)


```

```python
imagen.shape = (1, imagen.shape[0])
```

```python
new_model.predict(imagen, verbose=1)
```
